```python
# Specify the folder path containing the CSV files
folder_path = './ansel/data'

# Initialize an empty DataFrame to store the merged data
merged_data = pd.DataFrame()

# Iterate over each file in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        file_path = os.path.join(folder_path, file_name)
        # Read the CSV file
        df = pd.read_csv(file_path)
        # Merge the data with the existing DataFrame
        merged_data = pd.concat([merged_data, df], ignore_index=True)

# Specify the path for the output merged file
output_file_path = './ansel/merged_file.csv'

# Write the merged data to a new CSV file
merged_data.to_csv(output_file_path, index=False)
```

```python
# Set the food column as the index
df.set_index('nama', inplace=True)

# Generate unique IDs for each food item
df['FoodID'] = range(1, len(df) + 1)

# Set the food column as the index
df.set_index('FoodID', inplace=True)

# Get the unique categories
categories = df['kategori'].unique()

# Create a new DataFrame to store the category values
category_df = pd.DataFrame(index=df.index)

# Iterate over each category
for category in categories:
    # Create a new column for the category and set default value as 0
    category_df[category] = 0
    
    # Set the category value as 1 for foods belonging to that category
    category_df.loc[df[''] == category, category] = 1

# Save the category DataFrame to a new CSV file
category_df.to_csv('food_categories.csv')
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('./ansel/merged_file.csv')

# Specify the column you want to check for missing values
column_name = 'porsi'  # Ganti dengan nama kolom yang ingin Anda periksa

# Count the number of missing values in the specified column
missing_values_count = df[column_name].isnull().sum()

# Print the number of missing values
print("Jumlah data kosong pada kolom '{}': {}".format(column_name, missing_values_count))

for column in df.columns:
    # Check for missing values in the column
    missing_values = df[column].isnull().sum()
    
    # Check if there are any missing values in the column
    if missing_values > 0:
        print(f"Column '{column}' has {missing_values} missing values.")
        
        # Get the row indices where the missing values occur
        missing_rows = df[df[column].isnull()].index.tolist()
        print(f"Missing values occur in rows: {missing_rows}")
```

```python
import csv

def extract_column_data(csv_file, column_index, txt_file):
    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        column_data = [row[column_index] for row in reader]

    with open(txt_file, 'w') as file:
        for data in column_data:
            file.write(data + '\n')

# Example usage
csv_file = './ansel/merged_file.csv'  # Replace with your CSV file path
column_index = 0  # Replace with the index of the desired column (zero-based index)
txt_file = './ansel/column_data_nama.txt'  # Replace with the desired output .txt file path

extract_column_data(csv_file, column_index, txt_file)
```
